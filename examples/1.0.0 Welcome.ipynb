# COVID-19 Response Pangeo

**Welcome to the COVID-19 Response Pangeo!**

This Pangeo is provided free of charge by Microsoft Azure, the Met Office,
the Environmental Futures and Big Data Impact Lab and the EU European Regional Development Fund for analysing weather
data to explore links between the weather and the spread of COVID-19.



For more information see [our blog post](https://medium.com/informatics-lab/met-office-and-partners-offer-data-and-platform-for-covid-19-researchers-83848ac55f5f).

## What is this?

This is a cloud-based data processing platform where you can perform data analysis and visualisation work.  

This platform is a Pangeo instance. More information on Pangeo is available on the [Pangeo website](http://pangeo.io/).

This document is a Jupyter notebook. Jupyter notebooks are the default coding format on Pangeo. They allow you to write and execute code in a web browser, and mix code with rich text written in markdown format (what this is). Here's some more [information on Jupyter notebooks](https://jupyterlab.readthedocs.io/en/stable/user/notebook.html). There's also an introduction to notebooks provided in these examples, in `1.0.1 Notebook Intro.ipynb`.

## Exploring the data

The Met Office data available on this platform can all be found in the `datasets` folder in your homespace, which is visible in the _File Browser_ sidebar to the left. If the sidebar isn't visible, or is showing something that isn't a file browser, click the folder icon in the grey bar on the left of the screen.

In the dataset directory are three further directories for full details see the readme files in these directories:

* `open-data`: contains Met Office weather and air quality data for the UK and the world on daily (`data_nc_daily`) and hourly (`data_nc_hourly`) for the period from 01/01/20 to the present minus 48 hours. There is also a range of data that's been pre-processed to min/max/averages over different regions (`regional_subset_data`) and a directory of shapefiles that describe those regions (`shapefiles`). This data is avaliable for use under an open license.
* `non-commercial-data`: contains Met Office weather data for the UK for the years 2015 to 2019 (inclusive). This is avaliable for use under a non commercial license.
* `share`: a directory that's readable and writeable by all users of the platform. This makes it a good place to store data and notebooks you've created that you might wish to share with others.
